````python
import numpy as np
import math
x = np.linspace(1, 10, 200)

#1. Softplus Activation Function: A smooth approximation of ReLU is used in neural networks. f1(x) = log(1 + ex)
f1x = [math.log(1 + math.exp(x1)) for x1 in x]
print(f"f1x = {f1x}")

# 2. Log-Sum-Exp (LSE): A function often used in numerical stability for probability calculations. f2(x) = log Xn i=1 exi

f2x = math.log(np.sum([math.exp(x2) for x2 in x]))
print(f"f2x = {f2x}")

# 3. Gaussian Function: A commonly used function in probability and statistics. f3(x) = e**(−x2/2)
f3x = [math.exp(-x3**2/2) for x3 in x]
print(f"f3x = {f3x}")

#4. Sigmoid Function: A function used in logistic regression and neural networks. f4(x) = 1 1 + e−x
f4x = [(1/(1+math.exp(-x4))) for x4 in x]
print(f"f4x = {f4x}")

#5. Linear-Exponential Function: Combining linear and exponential components. f5(x) = xe−0.1x

f5x = [x5 * math.exp(-x5/10) for x5 in x]
print(f"f5x = {f5x}")
````

Q3. **Simple Linear Regression using numpy**
You are given a dataset as below. Your task is to implement a simple linear regression model
using the normal equation to compute the weight vector W. You will then evaluate the
model using the Mean Squared Error (MSE) and Mean Absolute Error (MAE) metrics.
Finally, you will visualize the actual vs. predicted values using Matplotlib.
You are required to use NumPy only to solve this problem.

Dataset
The given training dataset consists of 5 samples, where each sample has two features and a
bias term:

```
Xtrain =
1 1 2
1 2 5
1 3 7
1 4 10
1 5 12

, ytrain =
6
11
16
22
27

The test dataset consists of 3 samples:
Xtest =
1 6 14
1 7 16
1 8 19,

ytest =
33
38
43
```

````python
#1. Compute the weight vector W using the normal equation: W = (X^T X)^−1 X^T y
import numpy as np
import matplotlib.pyplot as plt

X_train = np.array([[1,1,2],[1,2,5],[1,3,7],[1,4,10],[1,5,12]])
X_test = np.array([[1,6,14],[1,7,16],[1,8,19]])

y_train = np.array([[6],[11],[16],[22],[27]])
y_test = np.array([[33],[38],[43]])


W = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train  # @ is used for matrix multiplication, .T is used for matrix Transpose
print(W)
````

````python
# Compute the predictions for the test set using: ypred = Xtest @ W
y_pred = X_test @ W
print(y_pred)

#Compute the Mean Squared Error (MSE) and Mean Absolute Error (MAE) for the predictions:

mse = np.mean((y_test - y_pred) ** 2)

mae = np.mean(np.abs(y_test - y_pred))

print(mse, mae)
````

4. **Plot the actual vs. predicted values using Matplotlib**:

• Use a scatter plot for actual vs. predicted values.

• Overlay a dashed red line to indicate perfect predictions (y = x).

• Label axes appropriately and provide a legend.


````python
import matplotlib.pyplot as plt

plt.figure(figsize = [8,6])

minval = min(np.min(y_test), np.min(y_pred))
maxval = max(np.max(y_test), np.max(y_pred))

plt.scatter(y_pred, y_test, color = 'b', label = "Actual vs predicted")

plt.plot([minval, maxval], [minval, maxval], color = 'r', label = "Perfect prediction")

plt.legend()

plt.show()
````

# 5.1 **Compute the probabilities**

z = X*train*W
